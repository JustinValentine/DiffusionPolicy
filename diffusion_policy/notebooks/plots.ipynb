```python
trajs =  np.load("../../outputs/cfm_traj.npy")
time_idx = 3
dx1, dy1 = np.diff(trajs[:, 0, time_idx, 0]), np.diff(trajs[:, 0, time_idx, 1])
plt.quiver(trajs[:-1, 0, time_idx, 1], trajs[:-1, 0, time_idx, 0], dy1, dx1, color='orange', scale=1, scale_units='xy', label='CFM', angles='xy', width=0.005)

trajs =  np.load("../../outputs/dp_traj.npy")
dx1, dy1 = np.diff(trajs[:, 0, time_idx, 0]), np.diff(trajs[:, 0, time_idx, 1])
plt.quiver(trajs[:-1, 0, time_idx, 1], trajs[:-1, 0, time_idx, 0], dy1, dx1, color='blue', scale=1, scale_units='xy', label='Diffusion Policy', angles='xy', width=0.005)
plt.xlabel("y")
plt.ylabel("x")
plt.title("Sampling Path")
plt.legend()
# plt.savefig('sampling_path.png', dpi=300, bbox_inches='tight')
```
